```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4]
dsa_marks = [90,92,93,95]
ml_marks=[91,93,94,95]
plt.plot(x,dsa_marks,label="dsa", color='g',marker='*')
plt.plot(x,ml_marks,label="ml",color='y')
plt.legend(loc="lower right")
plt.title("marks about students exam")
plt.xlabel("no of hours studied")
plt.ylabel("marks")
plt.xticks([1,2,3,4])
plt.yticks([90,92,93,95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

courses = ['os','cn','dbms','sql','ml','ai']
students = [2000,3000,4000,5000,6000,7000]
hatches = ['/','*','o','\\','|','-']
bar_colors = ['red','yellow','orange','green','pink','blue']

# Create bars without fixing color to 'g'
bars = plt.bar(courses, students, linewidth=2, edgecolor='black')

# Apply individual hatches and colors
for bar, hatch, color in zip(bars, hatches, bar_colors):
    bar.set_hatch(hatch)
    bar.set_color(color)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

courses = ['os','cn','dbms','sql','ml','ai']
students = [2000,3000,4000,5000,6000,7000]
hatches = ['/','*','o','\\','|','-']
bar_colors = ['red','yellow','orange','green','pink','blue']

# Create bars with black edges
bars = plt.bar(courses, students, linewidth=2, edgecolor='black')

# Apply hatch and color to each bar
for bar, hatch, color in zip(bars, hatches, bar_colors):
    bar.set_hatch(hatch)
    bar.set_facecolor(color)

plt.show()
```

```python

courses = ['os','cn','dbms','sql','ml','ai']
students = [2000,3000,4000,5000,6000,7000]
pie_colors = ['red','yellow','orange','green','pink','blue']
explode=[0,0,0,0,0,0.1]
plt.pie(students,labels=courses,explode=explode,colors=pie_colors,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create 3D axes
ax = plt.axes(projection="3d")

# Generate data points
x = np.arange(0, 100, 0.1)  # step size fixed
y = np.sin(x)
z = np.cos(x)

# Plot in 3D
ax.plot(x, y, z)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.axes(projection="3d")
x=np.random.randint(1,10,5)
y=np.random.randint(1,10,5)
z=np.zeros(5)

dx=np.ones(5)
dy=np.ones(5)
dz=np.random.randint(1,10,5)
ax.bar3d(x,y,z,dx,dy,dz)
plt.show()
```

```python
#3d plot surface

import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-5,5,0.25)
y=np.arange(-5,5,0.25)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))
ax=plt.axes(projection="3d")
ax.plot_surface(x,y,z,cmap="viridis")
plt.show()
```
