```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names
```

```python
penguins=sns.load_dataset("penguins")
sns.set_context("talk")
#penguins['species'].value_counts()
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue="island",style="sex")
```

```python
penguins=sns.load_dataset("penguins")
sns.swarmplot(data=penguins,x="species",y="body_mass_g" ,hue="island")
```

```python
#histogram plot
sns.histplot(data=penguins,x="body_mass_g",hue="island",multiple="stack")
```

```python
#regression plot
#sns.set_style("grid")
sns.regplot(data=penguins,x="body_mass_g",y="flipper_length_mm")
```

```python
#line plot
sns.lineplot(data=penguins,x="body_mass_g",y="flipper_length_mm")
```

```python
#jointplot
#barplot
sns.barplot(data=penguins,x="species",y="flipper_length_mm",hue="sex",palette="pastel")
```

```python
#countplot
sns.countplot(data=penguins,x="species",hue="island")
```

```python
#boxplot
sns.boxplot(data=penguins,x="species",y="body_mass_g")
```

```python
#kdeplot
sns.kdeplot(data=penguins,x="body_mass_g",hue="species",fill=True)
```

```python
#heatmap
columns=["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
sns.heatmap(data=penguins[columns].corr(),annot=True,vmin="-0.1")
```

```python
#rugplot
sns.rugplot(data=penguins,x="body_mass_g",y="flipper_length_mm")
```

```python
sns.pairplot(data=penguins)
```
